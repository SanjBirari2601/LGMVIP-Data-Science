# Beginner Task 4 - Image to Pencil Sketch with Python

Installing and Importing necessary packages

```python
!pip install opencv-python
import numpy as np
import cv2
from PIL import Image
from IPython.display import display
```

Loading and Displaying Image

```python
img= cv2.imread("E:\dog.jpg")
img= cv2.resize(img,(250,300))
RBG_img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(RBG_img))
```

Converting the original image into GreyScale Image 

```python
grey_img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grey_img))
```

Inverting the GreyScale Image

```python
inverted_grey_img= 255-grey_img
display(Image.fromarray(inverted_grey_img))
```

Changing to Negative Image

```python
Negative_img = cv2.bitwise_not(grey_img)
display(Image.fromarray(Negative_img))
```

Changing the Gaussian Blur Negative Image to Smooth Image

```python
smooth_img= cv2.GaussianBlur(Negative_img,(5,5),0,0)
display(Image.fromarray(smooth_img))
```

Changing to Pencil Sketch

```python
invert_blur=255- smooth_img
Pencil_sketch= cv2.divide(grey_img, invert_blur, scale= 250.0)
display(Image.fromarray(Pencil_sketch))
```

### Display of both original and Pencil Sketch

```python
display(Image.fromarray(RBG_img))
display(Image.fromarray(Pencil_sketch))
```
